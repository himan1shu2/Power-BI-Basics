Question 1 :Define Power BI and What are the key components of the Power BI ecosystem? Briefly explain:
  
   ● Power BI Desktop
   ● Power BI Service
● Power BI Mobile
● Power BI Gateway

Answer: Power BI is a business intelligence (BI) and data visualization tool by Microsoft that helps users collect, analyze, visualize, and share data. It transforms raw data from multiple sources into interactive reports and dashboards to support better decision-making.

Key Components of the Power BI Ecosystem
1. Power BI Desktop -
    - A free Windows-based application used to connect to data sources, transform data, and create reports.
    - Provides tools like Power Query for data preparation and DAX for calculations.
    - Reports created in Desktop are published to the Power BI Service.

2. Power BI Service -
     - A cloud-based platform (SaaS).
     - Used to publish, share, and collaborate on reports and dashboards.
     - Supports data refresh, security, and access control.
     - Accessible through a web browser.

3. Power BI Mobile -
      - A mobile app for Android, iOS, and Windows devices.
      - Allows users to view and interact with dashboards and reports anywhere.- Provides alerts and real-time access to business data.

4. Power BI Gateway -
     - A secure communication bridge between on-premises data sources (such as SQL Server, Oracle, or local files) and the Power BI Service in the cloud.
     - Allows scheduled refresh, live connections, and DirectQuery without moving on-premises data to the cloud.
     - Ensures data security by keeping data within the local network while only metadata and queries are sent to Power BI.
     - Commonly used in organizations where data cannot be stored directly in the cloud.

Answer: Comparison of Power BI Visuals -

       - Pie Chart -
          - Displays data as slices of a circle representing proportions of a whole.
          - Best when showing few categories.
          - Simple and easy to understand.

          When to prefer Pie Chart:
            - When you want a clear and simple view of percentage distribution.
            - Example: Showing market share of different brands.

           - Similar to a pie chart but has a hole in the center.

            - When you want to show percentages plus a key total or label.
            - Example: Displaying sales contribution by product category with total sales in the center.

 2. Bar Chart vs Column Chart

        - Displays data with horizontal bars.

 3. Comparing sales across different product names.

Question 3 : Explain the significance of:
  
  ● Star schema vs Snowflake schema
  
  ● Primary key vs Foreign key in relationships (Power BI)

Why is cardinality important?

Answer :
1. Star Schema vs Snowflake Schema :
      - Star Schema: Fact table → denormalized dimensions; faster queries; simpler in Power BI.
      
      - Snowflake Schema: Fact table → normalized dimensions; less storage, more joins; slower queries.

2. Primary Key vs Foreign Key :

     - Primary Key (PK): Unique identifier in a table (e.g., CustomerID).

     - Foreign Key (FK): Links to PK in another table; creates relationships for analysis.

3. Cardinality :

      - Defines relationship type: 1:1, 1:N, N:N.

      - Important for accurate calculations and filtering in Power BI.

      - Significance:

         - Determines how filters propagate between tables.
         - Impacts accuracy of calculations (like SUM, COUNT, AVERAGE).
         - Helps Power BI choose the best storage and performance model.
         - Incorrect cardinality can lead to duplicate counts or wrong results in visuals.

Question 4 : Differentiate between:
  
  ● Calculated column vs Measure

 Also, define Row context and Filter context with simple examples.

Answer : 1. Calculated Column vs Measure -

| Feature        | Calculated Column                                        | Measure                                              |
| -------------- | -------------------------------------------------------- | ---------------------------------------------------- |
| **Definition** | Column added to a table using a DAX formula             | Calculation done on the fly, not stored in table    |
| **Storage**    | Stored in the table                                      | Not stored; computed during query                   |
| **Evaluation** | Row by row                                               | Depends on filter context (aggregates)              |
| **Use case**   | Useful for **categorizing or transforming data** per row | Useful for **aggregations** like SUM, AVERAGE, COUNT|
| **Example**    | `Profit = Sales[Revenue] - Sales[Cost]`                 | `Total Profit = SUM(Sales[Profit])`                |


2. Row Context vs Filter Context
    - Row Context:

      - Applies when DAX formula works row by row.
      - Example:
           - In a calculated column: Profit = Revenue - Cost
           - Each row calculates its own Profit individually.

   - Filter Context:
        - Applies when DAX formula is influenced by filters in visuals or slicers.
        - Example:

           - Total Profit = SUM(Profit)
           - If you filter by “Region = East”, it sums only rows where Region = East.


Question 5: What is the difference between a report and a dashboard in Power BI?

Answer: Report vs Dashboard in Power BI -

| Feature         | Report                                           | Dashboard                                      |
| --------------- | ----------------------------------------------- | ---------------------------------------------- |
| **Definition**  | Multi-page view of data with visuals, tables, charts | Single-page summary of key metrics            |
| **Pages**       | Can have multiple pages                          | Only one page (tile-based)                     |
| **Interactivity** | Highly interactive (filters, slicers, drill-through) | Limited interactivity; mainly clickable tiles |
| **Data Sources** | Can use data from one or more datasets          | Combines visuals from multiple reports/datasets|
| **Purpose**     | For detailed analysis                            | For quick insights & monitoring KPIs           |
| **Example**     | Sales report showing product-wise and region-wise details | Dashboard showing total sales, top products, and revenue trends |
